```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
dir_ = 'D:/Datasets/tensor_board/'
```

```python
als_file = 'run_.-tag-ALS Loss.csv'
model_file = 'run_.-tag-Model Loss.csv'
auto_file = 'run_.-tag-Autoencoder Loss.csv'
```

```python
als_df = pd.read_csv(dir_  + als_file)
model_df = pd.read_csv(dir_  + model_file)
auto_df = pd.read_csv(dir_  + auto_file)
```

```python
stats = pd.DataFrame({'ALS': als_df.Value, 'Model': model_df.Value, 'Auto': auto_df.Value})
stats = stats.reset_index()
```

```python
stats.head(2)
```

```python
plt.yscale('log')
plt.title('ALS Loss', fontsize=14)
plt.xlabel('Epoch', fontsize=14)
plt.ylabel('Loss', fontsize=14)
plt.plot(stats.index, stats.ALS)
```

```python
plt.yscale('log')
plt.title('Model Loss', fontsize=14)
plt.xlabel('Epoch', fontsize=14)
plt.ylabel('Loss', fontsize=14)
plt.plot(stats.index, stats.Model)
```

```python
plt.yscale('log')
plt.title('Autoencoder Loss', fontsize=14)
plt.xlabel('Epoch', fontsize=14)
plt.ylabel('Loss', fontsize=14)
plt.plot(stats.index, stats.Auto)
```
